```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'airbnb.csv')
```

```python
type(df)
```

```python
df = df.drop('experiences_offered', axis=1)
```

```python
df.price = df.price.map(lambda x: float(x.replace('$','')))
```

```python
df_feats = ['id', 'accommodates', 'bathrooms', 'beds', 'price', 'weekly_price', 'monthly_price']
corr = df[df_feats].corr()
corr
```

```python
def norm_feat(series):
    return (series - series.mean())/series.std()
for feat in ['id', 'accommodates', 'bathrooms', 'beds', 'price', 'weekly_price', 'monthly_price']:
    df[feat] = norm_feat(df[feat])
df.describe()
```
